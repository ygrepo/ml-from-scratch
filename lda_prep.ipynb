# Latent Dirichlet Allocation (LDA)

### Dirichlet distribution.

```python
# Plot Dirichlet distributed points on a 2-simplex given parameter alpha.
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

alpha = (1.,1.,1.)
npts = 1000
s = np.random.dirichlet( alpha, npts )

fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111,projection='3d')

X = [s[i][0] for i in range(npts)]
Y = [s[i][1] for i in range(npts)]
Z = [s[i][2] for i in range(npts)]

ax.scatter(X,Y,Z)
ax.plot([0,1],[0,0],[1,0], color='black',linewidth=3)
ax.plot([1,0],[0,1],[0,0], color='black',linewidth=3)
ax.plot([0,0],[0,1],[1,0], color='black',linewidth=3)

ax.view_init(azim=45)
plt.show()
```
